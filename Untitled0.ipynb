```python
from statsmodels.tsa.stattools import adfuller, kpss

def test_stationarity(series, name):
    adf_result = adfuller(series.dropna())
    kpss_result = kpss(series.dropna(), regression='c', nlags='legacy')
    print(f"Feature: {name}")
    print(f"ADF p-value: {adf_result[1]} {'(Stationary)' if adf_result[1] < 0.05 else '(Non-stationary)'}")
    print(f"KPSS p-value: {kpss_result[1]} {'(Stationary)' if kpss_result[1] > 0.05 else '(Non-stationary)'}")

# low
# test_stationarity(df["close"], "close")
```

```python
import numpy as np

df['diff_close'] = df['close'].diff()
df.dropna(inplace=True)

# test_stationarity(df["diff_close"], "diff_close")
```

```python
plt.figure(figsize=(15, 8))
plt.plot(df['timestamp'], df['diff_close'], label='Differenced')
plt.legend()
plt.title('Differenced Price')
plt.show()
```

```python
def calculate_macd(df, fast_period, slow_period, signal_period):
    """
    MACD Metrics Calculation
    """
    ema_fast = df['close'].ewm(span=fast_period, adjust=False).mean()
    ema_slow = df['close'].ewm(span=slow_period, adjust=False).mean()
    macd = ema_fast - ema_slow
    signal = macd.ewm(span=signal_period, adjust=False).mean()
    macd_hist = macd - signal
    return macd, signal, macd_hist

# MACD
macd_parameters = [
    {'fast': 6, 'slow': 13, 'signal': 5},
    {'fast': 9, 'slow': 20, 'signal': 7},
    {'fast': 12, 'slow': 26, 'signal': 9},
    {'fast': 3, 'slow': 10, 'signal': 4},
    {'fast': 5, 'slow': 15, 'signal': 5},
    {'fast': 7, 'slow': 17, 'signal': 6},
    {'fast': 8, 'slow': 18, 'signal': 7},
    {'fast': 10, 'slow': 25, 'signal': 6},
    {'fast': 4, 'slow': 12, 'signal': 3},
    {'fast': 11, 'slow': 22, 'signal': 8}
]

df = df.sort_values('timestamp').reset_index(drop=True)

for params in macd_parameters:
    fast = params['fast']
    slow = params['slow']
    signal = params['signal']

    # Calculate MACD
    macd, signal_line, macd_hist = calculate_macd(df, fast, slow, signal)
    
    # add to DataFrame
    df[f'MACD_{fast}_{slow}_{signal}'] = macd
    df[f'Signal_{fast}_{slow}_{signal}'] = signal_line
    df[f'MACD_Hist_{fast}_{slow}_{signal}'] = macd_hist


# RSI
def compute_rsi(series, window):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Define RSI window sizes to try
rsi_windows = [5, 7, 14, 21]

# Calculate RSI for each window size
for window in rsi_windows:
    df[f'rsi_{window}'] = compute_rsi(df['close'], window=window)


# Define SMA & volume MA window sizes to try
ma_windows = [3, 5, 7, 10, 14, 20]

for window in ma_windows:
    df[f'sma_{window}'] = df['close'].rolling(window=window, min_periods=1).mean()
    df[f'volume_ma_{window}'] = df['volume'].rolling(window=window).mean()
    df[f'volume_ma_{window}'] = df['volume'].rolling(window=window).mean()

# price range and percent change
df['price_range'] = df['high'] - df['low']
df['percent_change'] = (df['close'] - df['open']) / df['open']


# lag features
lag_periods = [1, 2, 3, 4, 5]

features_to_lag = ['close']
for feature in features_to_lag:
    for lag in lag_periods:
        df[f'{feature}_lag_{lag}'] = df[feature].shift(lag)

features_to_lag_diff = ['diff_close']
for feature in features_to_lag_diff:
    for lag in lag_periods:
        df[f'{feature}_lag_{lag}'] = df[feature].shift(lag)

# money flow index
df['money_flow_multiplier'] = ((df['close'] - df['low']) - (df['high'] - df['close'])) / (df['high'] - df['low'])
df['money_flow_volume'] = df['money_flow_multiplier'] * df['volume']

df.dropna(inplace=True)
```

```python
split_ratio = 0.8
split_point = int(len(df) * split_ratio)

train = df.iloc[:split_point].reset_index(drop=True)
test = df.iloc[split_point:].reset_index(drop=True)
```

```python
from sklearn.utils import resample

positive = train[train['target'] == 1]
negative = train[train['target'] == 0]

print(f"Number of positive samples: {len(positive)}")
print(f"Number of negative samples: {len(negative)}")

negative_downsampled = resample(negative,
                                replace=False,
                                n_samples=len(positive),
                                random_state=42)

train_downsampled = pd.concat([positive, negative_downsampled])

print(f"Number of downsampled negative samples: {len(negative_downsampled)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# The distribution of the target variable -> I will do down-sampling after EDA
target_dist = train_downsampled['target'].value_counts(normalize=True)
print("Target Variable Distribution:")
display(target_dist)

plt.figure(figsize=(10, 6))
sns.countplot(data=train_downsampled, x='target')
plt.title("Target Variable Distribution")
plt.show()
```

```python
X_train = train_downsampled.drop(columns=['target', 'timestamp'])
y_train = train_downsampled['target']

X_test = test.drop(columns=['target', 'timestamp'])
y_test = test['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, f1_score
import matplotlib.pyplot as plt

model_tree = DecisionTreeClassifier(
    random_state=42,
    max_depth=5,
    min_samples_split=10,
    min_samples_leaf=5,
    max_features='sqrt'
)

model_tree.fit(X_train, y_train)

y_pred_tree = model_tree.predict(X_test)
y_pred_tree_prob = model_tree.predict_proba(X_test)[:, 1]

cnf_matrix_tree = confusion_matrix(y_test, y_pred_tree)
print("Confusion Matrix:")
print(cnf_matrix_tree)

f1_macro = f1_score(y_test, y_pred_tree, average='macro')
print("F1 Score (Macro):", f1_macro)


tree_fpr, tree_tpr, _ = roc_curve(y_test, y_pred_tree_prob)
auc_tree = roc_auc_score(y_test, y_pred_tree_prob)

plt.figure()
plt.plot(tree_fpr, tree_tpr, label=f'Decision Tree AUC {auc_tree:.3f}')
plt.plot([0, 1], [0, 1], 'r--')
plt.legend(loc=4)
plt.grid()
plt.show()
```

```python
feature_names = X_train.columns

importances = model_tree.feature_importances_

feature_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

print("Feature Importance:")
print(feature_importance_df)

plt.figure(figsize=(10, 18))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Decision Tree Feature Importances')
plt.gca().invert_yaxis()  
plt.show()
```

```python
X_train = X_train[feature_importance_df['Feature'][:25].values]
X_test = X_test[feature_importance_df['Feature'][:25].values]
```

```python
import lightgbm as lgb
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import f1_score, make_scorer

param_dist = {
    'num_leaves': [31, 50, 70],
    'max_depth': [5, 10, 15],
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 200, 300],
    'min_child_samples': [20, 30, 40],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

model_lgb = lgb.LGBMClassifier(
    objective='binary',
    random_state=42,
    class_weight='balanced'
)

cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)

f1_macro = make_scorer(f1_score, average='macro')

random_search = RandomizedSearchCV(
    estimator=model_lgb,
    param_distributions=param_dist,
    n_iter=20,
    scoring=f1_macro,
    cv=cv,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# 执行随机搜索
random_search.fit(X_train, y_train)

print("Best params:", random_search.best_params_)
print("Best F1 Score (Macro):", random_search.best_score_)


best_model = random_search.best_estimator_
best_model.fit(X_train, y_train)

y_pred = best_model.predict(X_test)
y_pred_prob = best_model.predict_proba(X_test)[:, 1]

from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# 混淆矩阵
cnf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cnf_matrix)

# F1 Score (Macro)
f1_macro_score = f1_score(y_test, y_pred, average='macro')
print("F1 Score (Macro):", f1_macro_score)

# ROC 曲线和 AUC
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
auc_score = roc_auc_score(y_test, y_pred_prob)

plt.figure()
plt.plot(fpr, tpr, label=f'LightGBM AUC {auc_score:.3f}')
plt.plot([0, 1], [0, 1], 'r--')
plt.legend(loc=4)
plt.grid()
plt.show()
```
